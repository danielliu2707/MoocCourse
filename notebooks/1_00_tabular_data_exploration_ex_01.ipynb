Imagine we are interested in predicting penguins species based on two of their
body measurements: culmen length and culmen depth. First we want to do some

What are the features? What is the target?

The data is located in `../datasets/penguins_classification.csv`, load it with
`pandas` into a `DataFrame`.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Write your code here.
penguins_df = pd.read_csv("../datasets/penguins_classification.csv")
```

```python
# Write your code here.
penguins_df['Species'].value_counts()
```

Plot histograms for the numerical features

```python
# Write your code here.
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,5))

# We can set the number of bins with the *bins* keyword argument.
for ax, col in zip(axes, penguins_df.columns):
    # ax is the keyword for the part of the overall figure in which a plot is drawn.
    # So, when you type "spx.plot(**ax=", you are declaring the values for that part of the figure.
    # Hence, plot on this part of the subplot:
    penguins_df[col].plot(kind="hist", ax=ax, bins=10, edgecolor="black")
    ax.set_title(f'{col} Distribution')
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')
```

Show features distribution for each class. Hint: use
[`seaborn.pairplot`](https://seaborn.pydata.org/generated/seaborn.pairplot.html)

```python
# Write your code here.
columns = ["Culmen Length (mm)", "Culmen Depth (mm)"]
target_column = "Species"

_ = sns.pairplot(
    data=penguins_df,
    vars=columns,
    hue=target_column,
    plot_kws={"alpha": 0.2},
    height=4,    # increase since axis labels overlapping
    diag_kind="hist",
    diag_kws={"bins": 30}
)
```

Looking at these distributions, how hard do you think it would be to classify
the penguins only using `"culmen depth"` and `"culmen length"`?


It looks feasible to classify penguins using just these two features. There appears to be a distinct boundary between Adelie, Gentoo and Chinstrap pengiuns.
